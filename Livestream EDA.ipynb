# LOADING THE DATASET

```python
column_names = ['Transaction_id','Transaction_time','Transaction_type' ,'Amount','Category','Mode','Narratives','Customer_id','customer_firstname','customer_lastName' ,'customer_gender','Date_of_birth','Salary','Account_Number','Status','Date_of_Creation','address_zipcode','address_state_address_City_address_Street','recipient id','duplicate']
df = pd.read_csv("savingtransactions 3.csv", header=None, names=column_names,index_col=False)
```

```python
df.head(100)
```

```python
def clean_and_split_address(address):
    if '@' in address:
        parts = address.split('@')
        cleaned_parts = [part.strip('-') for part in parts if part.strip('-')]
    else:
        parts = address.split('-')
        cleaned_parts = [part.strip() for part in parts]

    if len(cleaned_parts) >= 3:
        state = cleaned_parts[0]
        city = cleaned_parts[1]
        street = ' '.join(cleaned_parts[2:])
        return f"{state}, {city}, {street}"
    else:
        return "Invalid address format"

# Apply the clean_and_split_address function to the 'address_state_address_City_address_Street' column
df['corrected_address'] = df['address_state_address_City_address_Street'].apply(clean_and_split_address)

df.drop(columns=['address_state_address_City_address_Street'], inplace=True)

df.head()
```

#### some columns have mismatched entries so remodelling the dataframe

```python
# Create a duplicate column of 'recipient'
df['recipient_duplicate'] = df['recipient id'].astype(str)

# Remove all numeric values from the 'recipient_duplicate' column
df['recipient_duplicate'] = df['recipient_duplicate'].str.replace('\d+', '')

# Concatenate the new column with 'address_state_address_City_address_Street'
df['new_address'] = df['corrected_address'] + df['recipient_duplicate']

# Drop the 'recipient_duplicate' column
df.drop(columns=['recipient_duplicate','corrected_address'], inplace=True)

# Remove string values from 'recipient id' column
df['recipient id'] = df['recipient id'].str.extract('(\d+)', expand=False)

# Identify the misplaced values and move them back to 'recipient id'
df['recipient id'] = df.apply(lambda row: row['duplicate'] if pd.isna(row['recipient id']) else row['recipient id'], axis=1)

# Drop the 'duplicate' column
df.drop(columns=['duplicate'], inplace=True)

df.head(3)
```

#### some strings having mixed characters

```python
# Define a function to decode strings
def decode_string(s):
    try:
        # Attempt to decode the string using UTF-8 encoding
        decoded_str = s.encode('latin1').decode('utf-8')
        return decoded_str
    except UnicodeDecodeError:
        # If decoding fails, return the original string
        return s

# Apply the decoding function to 'customer_firstname' and 'customer_lastName' columns
df['customer_firstname'] = df['customer_firstname'].apply(decode_string)
df['customer_lastName'] = df['customer_lastName'].apply(decode_string)
```

```python
df.head(117)
```

#### spliting address column into state, city and street

```python
# Split 'new_address' column by '-'
split_address = df['new_address'].str.split(',', expand=True)

# Assign the split columns to 'state', 'city', and 'street' respectively
df['state'] = split_address[0]
df['city'] = split_address[1]
df['street'] = split_address[2]

# Drop the 'new_address' column
df.drop(columns=['new_address'], inplace=True)
```

#### Converting first letter of all words to capital

```python
# Define a function to capitalize each word in a string or capitalize single-word strings
def capitalize_text(text):
    if isinstance(text, str):
        words = text.split()
        if len(words) > 1:
            return ' '.join(word.capitalize() for word in words)  # Capitalize each word in multi-word string
        else:
            return text.capitalize()  # Capitalize single-word string
    else:
        return text  # Return non-string values unchanged

# Apply the function to all columns using applymap and update the DataFrame
df = df.applymap(capitalize_text)

df.head(2)
```

#### Normalizing the Gender Column

```python
# Define a function to map gender codes to full names
def map_gender(gender_code):
    if gender_code == 'F':
        return 'Female'
    elif gender_code == 'M':
        return 'Male'
    else:
        return gender_code  # Handle unexpected values (optional)

# Apply the function to the 'customer_gender' column using apply() and update the DataFrame
df['customer_gender'] = df['customer_gender'].apply(map_gender)
```

## Age

```python
from datetime import datetime 

# Get current date
current_date = datetime.now()

# Define a function to calculate age
def calculate_age(birthdate, current_date):
    age = current_date.year - birthdate.year - ((current_date.month, current_date.day) < (birthdate.month, birthdate.day))
    return age

# Apply the function to calculate age based on 'date_of_birth'
df['Age'] = df['Date_of_birth'].apply(lambda x: calculate_age(x, current_date))
```

```python
print("The skewness of customer age is:\n",df.groupby('Age')['Customer_id'].nunique().skew())
print("The kurtosis of the customer age is:\n",df.groupby('Age')['Customer_id'].nunique().kurt())
```

## Transaction_time

```python
# Extract month from 'transaction_time'
time=df.copy()
time['month'] = time['Transaction_time'].dt.month

# Define a function to categorize transactions into day or night
def categorize_time(hour):
    if 6 <= hour < 18:
        return 'Day'
    else:
        return 'Night'

# Extract hour from 'transaction_time'
time['hour'] = time['Transaction_time'].dt.hour

# Apply the function to categorize transactions into day or night
time['time_category'] = time['hour'].apply(categorize_time)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot the number of transactions in each month
sns.countplot(x='month', data=time, palette='viridis', ax=axes[0])
axes[0].set_title('Number of Transactions per Month')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Number of Transactions')

# Annotate counts on top of bars for month plot
for p in axes[0].patches:
    axes[0].annotate(f"{p.get_height()}", (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

# Plot the number of transactions in day and night time
sns.countplot(x='time_category', data=time, palette='muted', ax=axes[1])
axes[1].set_title('Number of Transactions in Day vs Night')
axes[1].set_xlabel('Time Category')
axes[1].set_ylabel('Number of Transactions')

# Annotate counts on top of bars for day vs night plot
for p in axes[1].patches:
    axes[1].annotate(f"{p.get_height()}", (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
```

More number of transactions happened in the month of march (42273)

More number of transactions are happening in day time (59324) than night time (42968)

## Transaction Type and category

```python
# Tabulation

tab = pd.crosstab(df['Category'], df['Transaction_type'])
print('the crosstab is:\n',tab)

# Create the countplot
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Category', hue='Transaction_type')
plt.title("Category vs. Transaction_type")
plt.xlabel("Category")
plt.ylabel("Count")
plt.legend(title='Transaction_type', loc='upper right')
plt.show()
```

```python
import scipy.stats as stats 

# Create a contingency table of 'Category' and 'Transaction_type'
df1 = df.dropna(subset=['Category', 'Transaction_type'])

contingency_table = pd.crosstab(df1['Category'], df1['Transaction_type'])

# Perform chi-square test of independence
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-Square Test of Independence Results:")
print(f"Chi-Square Statistic: {chi2}")
print(f"P-Value: {p_value}")
```

P value is greater than 0.05 indicating there is no realtion between transaction type and category
